```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib_inline import backend_inline
backend_inline.set_matplotlib_formats('retina', 'pdf')
# plt.rcParams['figure.figsize'] = [4, 4]
```

```python
def intensity(theta):
    const = 1
    return const * (np.sin(theta) * np.cos(theta)) ** 2


thetas = np.arange(-2 * np.pi, +2 * np.pi, 1e-4)

plt.polar(thetas, intensity(thetas), label=r'$I = \frac{1}{2 \mu_0 c}(\frac{\alpha \omega}{rc} sin\theta cos\theta)^2$')
plt.legend()
plt.show()
```
