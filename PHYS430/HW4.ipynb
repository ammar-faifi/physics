## Question 3

$N_A = 420$, $N_B = 180$, and $q= 100$

```python
N_A = 420; N_B = 180
N = N_A + N_B; q_total = 100

q_A = np.arange(0, q_total+1)
q_B = q_A[::-1]

omega_A = np.array([float(comb(q + N_A - 1, q)) for q in q_A])
omega_B = np.array([float(comb(q + N_B - 1, q)) for q in q_B])
omega_total = omega_A * omega_B
prob = omega_total / np.sum(omega_total)

plt.title('Multiplicity of the system $A$ as a function of $q_A$')
plt.xlabel('$q_A$')
plt.ylabel('$\Omega_A$')
plt.plot(q_A, omega_A)
plt.show()
```

```python
plt.title('Multiplicity of the system $B$ as a function of $q_A$')
plt.xlabel('$q_A$')
plt.ylabel('$\Omega_B$')
plt.plot(q_A, omega_B)
plt.show()
```

```python
plt.title('Multiplicity of the combined system as a function of $q_A$')
plt.xlabel('$q_A$')
plt.ylabel('$\Omega_t$')
plt.plot(q_A, omega_total)
plt.show()
```

```python
S_A = con.k * np.log(omega_A)
plt.title('Entropy of the system $A$ as a function of $q_A$')
plt.xlabel('$q_A$')
plt.ylabel('$S_A$')
plt.plot(q_A, S_A)
plt.show()
```

```python
S_B = con.k * np.log(omega_B)
plt.title('Entropy of the system $B$ as a function of $q_A$')
plt.xlabel('$q_A$')
plt.ylabel('$S_B$')
plt.plot(q_A, S_B)
plt.show()
```

```python
S_total = con.k * np.log(omega_total)
plt.title('Entropy of the combined system as a function of $q_A$')
plt.xlabel('$q_A$')
plt.ylabel('$S$')
plt.plot(q_A, S_total)
plt.show()
```

We see that the entropy of the combined systemm is maximum at $q_A = 70$.

```python
temp_A = np.diff(q_A) / np.diff(S_A)
temp_B = np.diff(q_B) / np.diff(S_B)

plt.title('Temperature of the system $A$ and $B$ as a function of $q_A$')
plt.xlabel('$q_A$')
plt.ylabel('$T$')
plt.plot(q_A[:-1], temp_A)
plt.plot(q_A[:-1], temp_B)
plt.show()
```
