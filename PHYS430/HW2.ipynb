### Problem 1.37

Compression of air to 1/20 of its original heat means $V_f = V_i / 20$, for air we have $f=5$ and $T_i = 300$ K. So from equation
$$
    V_f T_f^{f/2} = V_i T_i^{f/2}
$$
Then final temperature will be
$$
    T_f = T_i \left( \frac{V_i}{V_f} \right)^{2/f} = 
    (300 \text{ K}) (20)^{2/5} = 993 K
$$
