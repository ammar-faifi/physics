# Stating the Equations

The lagrangian is defined as $L \equiv T - U$, where $T$ is the kinetic energy for the considred mass, and $U$ is the potential energy on due to forces on the mass. For the $j$th particle, we have $T_j$ and $U_j$ as 
$$T_j = \frac{1}{2} m \dot{x_j}^2 \qquad U_j = \frac{1}{2} k x_j^2 $$

We relize that we have $N$ particles connected by $N+1$ springs. Start analayzing by assuming all the particles have shifted from their equilibrium a distance $x_j$ to the right.


## First Atom
Consider the free body diagram for the first atom we find its potential as 
$$ U_1 = \frac{k}{2}k x_1^2 + \frac{k}{2} (x_2^2 - x_1^2) = \frac{k}{2} (x_2^2 - 2x_1x_2 + 2x_1^2) $$
And the kinetic energy is
$$ T_1 = \frac{m}{2} \dot{x_1}^2 $$
Then the Lagrangian is 
$$ L_1 = T_1 - U_1 = \frac{m}{2} \dot{x_1}^2 - \frac{k}{2} (x_2^2 - 2x_1x_2 + 2x_1^2) $$

Since we take the partial derivatives with respect to $x_1$ it is enough to use $L_1$, then the Lagrange equation is 
$$ \frac{\partial L_1}{\partial x_1} - \frac{d}{dt}\frac{\partial L_1}{\partial \dot{x_1}} = 0 $$
So, 

$$ - \frac{k}{2} (-2 x_2 + 4x_1) - \frac{d}{dt} m \dot{x}_1 = 0 $$
$$ - k (- x_2 + 2 x_1) -  m \ddot{x}_1 = 0 $$
$$ \ddot{x}_1 + 2\omega_0^2 x_1 - \omega_0^2 x_2 = 0 $$
where, $\omega_0^2 \equiv k/m$

## Last Atom; $N$
Consider the last atom at $j=N$, its previous spring streached by a distance $x_N - x_{N-1}$, and its next spring compressed by a distance $x_N$. The Lagrangian is 
$$ L_N = \frac{m}{2} \dot{x}_N^2 - \left(\frac{k}{2} x_N^2 + \frac{k}{2}(x_N - x_{N-1})^2 \right) $$
We can use the last equation and just replace $x_N$ with $x_N$ and $x_{N+1}$ with $x_{N+1}$, that is, 
$$ \ddot{x}_N + 2\omega_0^2 x_N - \omega_0^2 x_{N-1} = 0 $$

## Middle Atoms
Again, we analyze the free body diagram for a given in between atom to its potential energy. We find that the Lagrangian is,
$$ L_j = T_j - U_j = \frac{m}{2} \dot{x}_j^2 - \frac{k}{2} \left[(x_{j} - x_{j-1})^2 + (x_{j+1} - x_j)^2\right] $$

Or 
$$ L_j = \frac{m}{2} \dot{x}_j^2 - \frac{k}{2} \left[ 2x_j^2 - 2x_j x_{j-1} - 2x_j x_{j+1} +x_{j-1}^2 + x_{j+1}^2 \right] $$

Then the equation of motion is,
$$ \frac{\partial L}{\partial x_j} - \frac{d}{dt}\frac{\partial L}{\partial \dot{x_j}} = 0 $$

Then,
$$ m \ddot{x} + k [ 2 x_j - x_{j-1} - x_{j+1} ] = 0 $$
$$ \ddot{x} +  2\omega_0^2 x_j - \omega_0^2 x_{j-1} - \omega_0^2 x_{j+1} = 0 $$

## Generalization of Lagrangian
$$ L = \sum_{j=1}^N (T_j - U_j) = 
\sum_{j=1}^N \left(\frac{m}{2} \dot{x_j}^2 - \frac{k}{2} (x_{j+1}^2 - 2x_j x_{j+1} + 2x_j^2)\right)$$ 

The Lagrange equation is
$$ \frac{\partial L}{\partial x_j} - \frac{d}{dt}\frac{\partial L}{\partial \dot{x_j}} = 0, \qquad j = 1, 2, 3, \dots, N $$

# The Eigenmodes

We write the equation of motion in matrix form as, $\ddot{X} + A X = 0 $, a second-order system.
$$
\begin{bmatrix}
\ddot{x}_1 \\ \ddot{x}_2 \\ \vdots \\ \ddot{x}_N
\end{bmatrix} +
\begin{bmatrix}
2 \omega_0^2 & -\omega_0^2 & 0 & & 0 & 0 & 0 & & 0 & 0 & 0 \\
-\omega_0^2 & 2 \omega_0^2 & - \omega_0^2 & & 0 & 0 & 0 & & 0 & 0 & 0 \\
0 & - \omega_0^2 & 2 \omega_0^2 & & 0 & 0 & 0 & & 0 & 0 & 0 \\
\\
0 & 0 & 0 & & 2 \omega_0^2 & - \omega_0^2 & 0 & & 0 & 0 &0 \\
0 & 0 & 0 & & - \omega_0^2 & 2 \omega_0^2 & - \omega_0^2 & & 0 & 0 & 0 \\
0 & 0 & 0 & & 0 & - \omega_0^2 & 2\omega_0^2 & & 0 & 0 & 0 \\
 & \vdots \\
0 & 0 & 0 & & 0 & 0 & 0 & & 2 \omega_0^2 & -\omega_0^2 & 0 \\
0 & 0 & 0 & & 0 & 0 & 0 & & -\omega_0^2 & 2 \omega_0^2 & - \omega_0^2 \\
0 & 0 & 0 & & 0 & 0 & 0 & & 0 & - \omega_0^2 & 2 \omega_0^2
\end{bmatrix}
\begin{bmatrix}
x_1 \\ x_2 \\ \vdots \\ x_N
\end{bmatrix} = 0
$$


We assume a solution of form $\vec{x}(t) = \vec{v} e^{\alpha t}$, where $\vec{v}$ is constant vector.  Substitute it in the matrix equation, we get 
$$ \alpha^2 \vec{v} e^{\alpha t} + A \vec{v} e^{\alpha t} = 0 $$
$$ A \vec{v} - (- \alpha^2) \vec{v} = 0 $$
Then, $\vec{x}(t)$ is a solution of the matrix Eq. if and only if $-\alpha^2 = \lambda$ is an eigenvalue of the matrix $A$, and $\vec{v}$ is an associated eigenvector(eigenmode). Since we are considering real-valued solutions, eigenvalues should be postive real numbers. We take $-\alpha^2 = \omega^2 > 0$, and $\alpha = \pm \omega i$. The solution is in form of $\vec{x}(t) = \vec{v} e^{i \omega t}$ Then we get, 

$$ A \vec{v} - \omega^2 \vec{v}= \vec{0} \qquad \Rightarrow \qquad (A - \omega^2 I) \vec{v}= \vec{0} $$
We define $A - \omega^2 I \equiv D_N$. To find nontrivial solutions we make $|D_N|=0$.

```python
# import needed packages
import numpy as np
import scipy.linalg

# graph settings
import matplotlib.pyplot as plt
from matplotlib_inline import backend_inline
plt.rcParams['figure.figsize'] = [13, 7]
backend_inline.set_matplotlib_formats('svg', 'pdf')
```

Now find the number of eigenvalues less thatn $\omega$, called $N(\omega)$. And the **density of states** is defined as 
$$ \rho(\omega) = \frac{dN}{d\omega} $$
Or 
$$ \rho(\omega) = \frac{N(\omega + d\omega) - N(\omega)}{d\omega} $$
where $d\omega$ is small energy step to be fixed.

# Plotting $\rho$ vs. $\omega$
Let $\omega^2 = 1$.

```python
# define the counting function of eigenvalues
def N(omega, eig):
    counter = 0
    for i in np.nditer(eig):
        if i < omega:
            counter += 1
    return counter

# define the density function
def rho(omega, domega, eig):
    return ( N(omega + domega, eig) - N(omega, eig) ) / domega

domega = 1.5e-1   # small step in omega
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)  # create 4 axes in 1 figure
Ns = np.array([500, 1000, 2000, 3000]).reshape((2, 2))   # number of atoms to try

for i in range(2):
    for j in range(2):
        n = Ns[i][j]                        # choose a value for N
        A = np.diagflat(2*np.ones(n), k=0)  # NxN matrix with diagonal elements equal 2
        A -= np.diagflat(np.ones(n-1), k=1) # substract 1 from the upper diagonal
        A -= np.diagflat(np.ones(n-1), k=-1)# substract 1 from the lower diagonal

        eigvals = np.linalg.eigvals(A)      # find the eigenvalues
        om_values = np.sqrt(np.linalg.eigvals(A))   # elementwise sqrt
        om_values = np.concatenate((-om_values, om_values)) # fabricate the negative values
        # plot
        axs[i][j].plot(eigvals, [rho(omega, domega, eigvals) for omega in np.nditer(eigvals)], '.')
        axs[i][j].set_title(f'DOS of $\omega^2$, for $N={n}$')
plt.show()
```

We notice at the edges as like the DOS approaches a max values. These edges are limiting the range possible for the eigenvalues to be exist.That is, outside this range is not for the system to occupy.
We can deduce that the first edge and the second one are 0 and 4, resprectively.Hence, $\omega_1=0$, and $\omega_2=2$

# Loop Through $N$
Now we will loop through some values of $N$ to see the eigenvalurs distribution

```python
y = np.array([])
x = np.array([])

for N in np.arange(1, 40+1, 1):
    A = np.diagflat(2*np.ones(N), k=0)  
    A -= np.diagflat(np.ones(N-1), k=1) 
    A -= np.diagflat(np.ones(N-1), k=-1)
    eig = np.linalg.eigvals(A)

    y = np.append(y, eig)
    x = np.append(x, np.repeat(N, eig.size))    # repeat the same N as many as eigenvalues 

plt.rcParams['figure.figsize'] = [13, 4]
plt.plot(x, y, '.')
plt.show()
```
